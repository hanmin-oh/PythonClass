melt() 메소드

데이터프레임을 깔끔한 데이터로 정리하는데 사용하는 메소드로 지정한 열에 대해서 데이터를 모두 행으로 정리한다.  
melt(데이터프레임, id_vars, value vars, var_name, value_name) 
id_vars : 위치를 그대로 유지할 열 이름, 필수  
value_vars : 행으로 위치를 변경할 열 이름, 생략 시 전체 열을 대상으로 한다.  
var_name : value_vars로 위치를 변경한 열 이름을 지정한다. 생략시 variable가 지정된다.  
value_name : value_vars로 위치를 변경한 열의 데이터가 나타나는 열의 이름을 지정한다. 생략시 value가 지정된다. 

```python
# 소득 정보가 열을 구성하고 있다. => 소득 정보 열을 행 데이터로 옮기려고 한다.
# id_vars 속성으로 지정한 열(religion)을 제외한 나머지 열(소득 정보, <$10k, 10-20k, ... ,> 150k , Don't know/refused)이 variable 열로 소득 정보 열의 데이터는 value열로 정리된다.
# value_vars 속성을 생략하면 모든 열을 대상으로 메소드가 실행된다.
# 이러한 과정을 'religion'열을 고정하여 피벗햇다.
pew_long = pd.melt(pew, id_vars='religion')
pew_long
```

```python
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k')
pew_long
```

```python
# 피벗할 열이 2개 이상이라면 피벗할 열을 리스트로 묶어서 value_vars 속성으로 전달하면 된다
pew_long = pd.melt(pew, id_vars='religion', value_vars=['<$10k','>150k'])
pew_long
```

```python
# var_name 속성을 이용해서 variable로 표시되던 열 이름을 '연봉'으로 변경하고 value_name 속성을 이용해서 value로 표시되던 열 이름을 '인원수'로 변경한다.
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k' , var_name='연봉', value_name='인원수')
pew_long.columns = ['종교', '연봉' ,'인원수']
pew_long
```

```python
ebola = pd.read_csv('./data/country_timeseries.csv', sep=',')
print(type(ebola))
```

```python
ebola_short = ebola.loc[:, ['Date','Day','Cases_Guinea' , 'Cases_Liberia' , 'Deaths_Guinea', 'Deaths_Liberia']]
ebola_short
```

```python
# date 열과 day열을 고정하고 나머지 열을 행으로 피벗한다.
ebola_short = pd.melt(ebola_short, id_vars=['Date', 'Day'])
ebola_short
```

```python
# split() 메소드에 '_' 구분자로 지정하면 Cases_Guinea를 Cases와 Guinea로 분리할 수 있다. 
print(ebola_short.variable) # <class 'pandas.core.series.Series'>
print('=' * 80) 
# 시리즈 뒤에 '.'을 찍고 str을 붙여주면
print(type(ebola_short.variable.str)) # <class 'pandas.core.strings.accessor.StringMethods'> 
variable_split = ebola_short.variable.str.split('_') 
print(variable_split)  #<class 'pandas.core.series.Series'>
print(variable_split[0]) # <class 'list'>
print(variable_split[0][0])
print(variable_split[0][1])
# variable_split 여기에 str을 한번 더 붙여준 후 인덱싱을 하면ㅋ 리스트로 원하는 데이터 추출 가능
print('='* 80)
print(variable_split.str[0])
print(variable_split.str[1])
```

```python
status_values = variable_split[0][0]
print(status_values)
```

```python
country_values = variable_split.str.get(1)
print(country_values)
```

```python
print(ebola_short.values)
ebola_short['status'] = variable_split.str[0]
ebola_short['country'] = variable_split.str[1]
ebola_short
```

```python
# split() 메소드에 expand 옵션을 True로 지정하면 split() 메소드가 실행한 결과를 데이터프레임으로 만든다.
variable_split = ebola_short.variable.str.split('_', expand=True)
# print(variable_split)
# variable_split
variable_split.columns = ['status2', 'country2']
variable_split
```

```python
ebola_parsed = pd.concat([ebola_short, variable_split], axis=1)
ebola_parsed
```
