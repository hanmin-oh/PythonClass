```python
import requests

targetSite = 'https://eggdrop.co.kr/api/store.php'
params = {
    'sido': '',  # 시도명
    'gugun': '',  # 구군명
    'storename': '',  # 매장명 (필터링을 위한 검색어)
    'filter1': '',    # 필터 1
    'filter2': '',    # 필터 2
    'filter3': '',    # 필터 3
    'filter4': '',    # 필터 4
    'filter5': ''     # 필터 5
}

response = requests.get(targetSite, params=params)
eggdrop = response.json()
eggdrop_df = pd.json_normalize(eggdrop, 'data')
print(type(eggdrop_df)) # <class 'pandas.core.frame.DataFrame'>
print(len(eggdrop_df))
print(eggdrop_df.columns)
# print(eggdrop)
# print(type(eggdrop))
eggdrop_df_map = eggdrop_df[
    [
   'seq', 'storecode', 'title_kr', 'phone', 'operatingtime',
       'sido', 'gugun', 'addr1','latitude',
       'longitude'
    ]
]
eggdrop_df_map
eggdropMap = folium.Map(location=[36.51156 , 127.249449], zoom_start=17)
for index, data in eggdrop_df_map.iterrows():
    popup = folium.Popup('{}. {} -{}'.format(index+1 , data['title_kr'], data['addr1']) , max_width=300)
    folium.Marker(location=[data['latitude'],data['longitude']] , popup=popup).add_to(eggdropMap)
eggdropMap.save('./eggdrop전국.html')
eggdropMap
```
