```python
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
# requests 모듈의 get() 메소드로 타겟 사이트의 정보를 읽어온다.
request = requests.get(targetSite)
print(request)
```

```python
html = request.text
print(html)
```

```python
soup = BeautifulSoup(html , 'html.parser')
print(soup)
```

```python
topices = soup.findAll('div' , {'class' , 'tcb-flex-col'})
print(len(topices))
print('=' * 80)
subjects = []
i = 0
for topice in topices :
    # print(topice)
    chapter = topice.findAll('a')
    # print(len(chapter))
    for a in chapter:
        i += 1
        print('{:2d}. {} '.format(i, a.text))
        subjects.append(a.text)
```

```python
print(subjects)
```
